```python
import pandas as pd
import numpy as np

exam_data= {'name': ['Anastasia','Dima','Katherine','James','Emily','Micheal','Matthew','Laura','Kevin','Jonas'],
           'score': [12.5,9,16.5, np.nan,9,20,14.5, np.nan,8,19],
           'attempts': [1,3,2,3,2,3,1,1,2,1],
           'qualify': ['yes','no','yes','no','no','yes','yes','no','no','yes']}
labels=['a','b','c','d','e','f','g','h','i','j']

df=pd.DataFrame(exam_data, index=labels)
print(df)



```

```python
print("Original rows:")
print(df)
print("\nChange the name 'James' to 'Suresh':")
df['name'] = df['name'].replace('James','Suresh')
print(df)
```

```python
selected_rows = df[(df['attempts'] < 2) & (df['score'] > 15)]
print(selected_rows)
```

```python
print("\nOriginal data frame:")
print(df)
print("\nChange the score in row 'd' to 11.5:")
df.loc['d', 'score'] = 11.5
print(df)
```

```python
print("\nSum of the examination attempts by the students:")
print(df['attempts'].sum())


```

```python
print("\nMean score for each different student in data frame:")
print(df['score'].mean())
```

```python
print("Original rows:")
print(df)
print("\nAppend a new row:")
df.loc['k'] = [1, 'Rahul', 'yes', 25.5]
print("Print all records after insert a new record:")
print(df)
print("\nDelete the new row and display the original rows:")
df = df.drop('k')
print(df)

```

```python
#Rows where scores is missing
print(df[df['score'].isnull()])
```

```python
#Sort the dataframe name in descending order and score in ascending order
df = df.sort_values(by=['name', 'score'], ascending=[False, True])
print(df)
```

```python
df['qualify'] = df['qualify'].map({'yes': True, 'no': False})
print(df)
```

```python
city=['Delhi','Mumbai', 'Banglore','Kolkata','Chennai','Bihar','Lucknow','Assam', 'MadhyaPradesh','Hyderabad']
df['city'] = city
print(df)
```
